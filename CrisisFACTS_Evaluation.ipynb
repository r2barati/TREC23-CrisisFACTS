```python
# Importing required libraries
import json
import datetime
import pandas as pd

# Reading the JSON file
file_path = "CrisisFACTs-2022.facts.json"
with open(file_path, 'r') as file:
    data = json.load(file)

# Prompt for the numerical part of the event ID
event_num = input("Please enter the numerical part of the event ID (e.g., 001): ")
event_id = f"CrisisFACTS-{event_num}"

# Extracting facts related to the given event ID
event_facts = [event for event in data if event["eventID"] == event_id]
event_name = event_facts[0]["event"] if event_facts else "N/A"

# Load the queries data
queries_file_path = 'CrisisFACTs-2022-v3.queries.json'
queries_df = pd.read_json(queries_file_path)

# Extract the related queries
related_queries = queries_df[queries_df['eventID'] == event_id]['queries'].iloc[0]

# Print the queries related to the event
print(f"\nQueries related to the event {event_id} - {event_name}:\n")
for i, query in enumerate(related_queries, 1):
    print(f"{i}. {query['query']}")

# Printing facts related to the given event ID
if event_facts:
    print(f"\nFacts related to the event {event_id} - {event_name}:")
    for event_fact in event_facts:
        facts_by_request = event_fact["factsByRequest"]

        # Creating a list of tuples containing request ID, date string, and facts
        requests_and_facts = [(request_id, [request['dateString'] for request in event_fact["summaryRequests"] if request['requestID'] == request_id][0], facts)
                              for request_id, facts in facts_by_request.items()]

        # Sorting the list based on the date string
        requests_and_facts.sort(key=lambda x: x[1])

        print("\nFacts by Request:")
        for request_id, date_string, facts in requests_and_facts:
            print(f"\nRequest ID: {request_id} (Date: {date_string})")
            for i, fact in enumerate(facts, 1):
                date_unix = fact['dateUnix']
                date_time = datetime.datetime.fromtimestamp(date_unix / 1000).strftime('%Y-%m-%d %H:%M:%S')
                print(f"  {i}.  {fact['fact']}")
else:
    print(f"No facts found for the event {event_id} - {event_name}.")
```

To pass the evaluations in the CrisisFACTS project, you need to perform well in the two evaluation metric sets: ROUGE-based Summarization Against Daily Summaries (Metric Set 1) and Individual Fact-Matching Between Runs and Manual Fact Lists (Metric Set 2). Here's an overview of each metric set and how to perform well:

**Metric Set 1 - ROUGE-based Summarization Against Daily Summaries:**
- In this metric set, your system's output will be aggregated into a "summary" of each day's developments for the disaster events.
- This summary will be compared against extant summaries using evaluation metrics such as ROUGE-x, BERTScore, and similar measures.
- Gold-standard summaries for evaluation include NIST-based assessments of the events and related summaries, as well as summaries extracted from the ICS209 reports from FEMA when available.
- To perform well, your system's summaries should be concise, accurate, and relevant to the information needs defined in the user profiles. The system should effectively identify critical new developments for each day of the disaster event.

**Metric Set 2 - Individual Fact-Matching Between Runs and Manual Fact Lists:**
- NIST assessors will construct a set of facts for each event-day pair by pooling facts from all participant runs.
- The assessors will then match the output from participant runs to the gold-standard facts to assess comprehensiveness and redundancy ratio, acting as proxies for recall and precision, respectively.
- Comprehensiveness is a recall-oriented metric that measures how well your system captures the relevant facts from the gold-standard fact set.
- Redundancy measures the number of submitted facts that match the gold-standard facts, which indicates the precision of your system in identifying relevant information without duplication.
- To perform well, your system should provide comprehensive and non-redundant facts that closely match the gold-standard facts produced by NIST assessors.

Overall, to succeed in the CrisisFACTS project, your system should demonstrate effective fact extraction, summarization, and matching capabilities. It should provide accurate and relevant information that aligns with the information needs defined in the user profiles. The system should also avoid redundancy and focus on delivering critical new developments for each disaster event day. Regularly testing and refining your system based on feedback from the evaluations will help improve its performance and achieve better results.

## Submission
To achieve the submissions for the CrisisFACTS project, follow these steps:

1. **Automatic Runs:**
   - If you are submitting an automatic run, your system should be designed to work without human intervention once it is started and provided with the task inputs.
   - Indicate in your submission that it is an automatic run to inform the organizers about the nature of your system.

2. **Manual Runs:**
   - If you are submitting a manual run, it means that a person (human) manually changes, summarizes, or re-ranks queries, the system, or the system's lists of facts.
   - Simple bug fixes that address only format handling do not result in manual runs, but if you make significant changes that involve manual intervention, you should describe those changes.

3. **Submission Format:**
   - Prepare your submission in the newline-delimited JSON file format as specified by the organizers.
   - Each entry in the JSON file should contain the fields outlined in the "System Output" section, which includes "requestID," "factText," "unixTimestamp," "importance," "sources," "streamID," and "informationNeeds."

4. **Submission File Structure:**
   - Each submission file corresponds to a single submitted run and should include all event-day pairs for all events.
   - Make sure to include the submission's runtag in the filename, as this will help in identifying and organizing the submissions.

5. **Submit through NIST System:**
   - Go to the NIST submission system at trec.nist.gov to upload your submission.
   - Ensure that your submission passes validation, as runs that do not pass validation will be rejected outright.

6. **Specify Run Details:**
   - During the submission process, you will be asked to specify certain details related to your run, such as:
     - Whether the run is manual or automatic (choose the appropriate option based on your system's nature).
     - Whether your system uses extant TREC-IS labels for Twitter data (if applicable).
     - Whether the summaries built are abstractive or extractive (choose the appropriate option based on your system's approach to summarization).

7. **Example Submissions:**
   - If available, refer to the example submissions provided in the "Output Examples" to understand the format and structure expected in the submission files.

Ensure that your submission adheres to the guidelines and format specified by the organizers. It is essential to thoroughly test your system before making the final submission to ensure its accuracy and compliance with the evaluation metrics. Regularly check the official CrisisFACTS website and communication channels for updates and announcements related to the submission process and deadlines.
